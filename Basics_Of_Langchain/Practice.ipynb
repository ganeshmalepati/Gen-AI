```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
text = TextLoader('speech.txt')
docs = text.load()
docs

speech = ""
with open('speech.txt') as f:
    speech=f.read()

loader = CharacterTextSplitter(chunk_size=500, chunk_overlap=20)
docs = loader.create_documents([speech])
print(docs)
print(docs[0])
print(docs[1])
```

```python
# How to split JSON data into chunks
import json
import requests
from langchain_text_splitters import RecursiveJsonSplitter
json_data = requests.get("https://api.smith.langchain.com/openapi.json").json()
json_data
```

```python
json_splitter = RecursiveJsonSplitter(max_chunk_size=100)
final_json_data=json_splitter.split_json(json_data)
# final_json_data

result = json_splitter.split_text(json_data)
result
print(result[0])
print(result[1])
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
query = "This session is about the OPENAI Embedding techniques"
res = embeddings.embed_query(query)
res
```

```python
from langchain_community.document_loaders import WikipediaLoader
docs = WikipediaLoader(query="Generative AI", load_max_docs=2)
res = docs.load_and_split()
print(res)
```
